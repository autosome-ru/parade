# PARADE prediction tutorial

In this tutorial, we will use PARADE predictor to estimate the stability of arbitrary 3'UTR sequences.

### Requirements

This code requires the following core Python modules:
* NumPy, Pandas
* PyTorch (preferably with CUDA support, if you have GPU)
* PyTorch Lightning (for PARADE wrappers)

```python
import sys
import numpy as np
import pandas as pd

import torch
torch.cuda.is_available()
```

### Import PARADE modules

```python
sys.path.append('../predictor/model/')
from predict import predict_sequences
```

Let's generate several random sequences to test the model.

The model was trained on 186nt long mRNA 3'UTR sequences (`T` stands for uracil).

The predictor yields reasonable assessments for sequences of similar length (~ 150-250nt), but with much shorter or longer 3'UTRs it is suggested to apply the predictor the the 5' region of the 3'UTR sequence.

```python
def generate_random_sequences(n: int, length=186, alphabet=['A', 'T', 'G', 'C']):
    return [''.join(np.random.choice(alphabet, size=(length,), replace=True)) for _ in range(n)]
```

```python
sequences = generate_random_sequences(10)
sequences
```

To predict the activity of an arbitrary sequence, pass it to the `predict_sequences` function.

Supply it only with the sequences of strictly equal length.

```python
result = predict_sequences(sequences, utr_type='stability', device='cpu')
result
```

To use a GPU, specify `device='cuda'`:

```python
result = predict_sequences(sequences, utr_type='stability', device='cuda')
result
```

A Pandas `Series` object is returned.

Save the result with Pandas

```python
result.to_csv('scores_stability.csv')
```
