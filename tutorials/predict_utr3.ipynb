# PARADE prediction tutorial

In this tutorial, we will use PARADE predictor to estimate the activity of arbitrary sequences.

### Requirements

This code requires the following core Python modules:
* NumPy, Pandas
* PyTorch (preferably with CUDA support, if you have GPU)
* PyTorch Lightning (for PARADE wrappers)

```python
import sys
import numpy as np
import pandas as pd

import torch
torch.cuda.is_available()
```

### Import PARADE modules

```python
sys.path.append('../predictor/model/')
from predict import predict_sequences
```

Let's generate several random sequences to test the model.

The model was trained on 240nt long mRNA 3'UTR sequences (`T` stands for uracil).

The predictor yields reasonable assessments for sequences of similar length (~ 200-300nt), but with much shorter or longer 3'UTRs it is suggested to apply the predictor the the 5' region of the 3'UTR sequence.

```python
def generate_random_sequences(n: int, length=240, alphabet=['A', 'T', 'G', 'C']):
    return [''.join(np.random.choice(alphabet, size=(length,), replace=True)) for _ in range(n)]
```

```python
sequences = generate_random_sequences(10)
sequences
```

To predict the activity of an arbitrary sequence, pass it to the `predict_sequences` function.

Supply it only with the sequences of strictly equal length.

```python
result = predict_sequences(sequences, utr_type='utr3', device='cpu')
result
```

To use a GPU, specify `device='cuda'`:

```python
result = predict_sequences(sequences, utr_type='utr3', device='cuda')
result
```

A Pandas `DataFrame` object is returned. PARADE was trained to predict 3'UTR activities for 6 cell lines:
* MDA-MB-231 (`c1`)
* HepG2 (`c2`)
* Jurkat (`c4`)
* SW480 (`c6`)
* PA-1 (`c13`)
* NALM6 (`c17`)

You can also select particular cell lines for your query if you are doing a massive _in silico_ screening:

```python
result = predict_sequences(sequences, utr_type='utr3', device='cpu', cell_line_codes=['c1', 'c6'])
result
```

Save the result with Pandas

```python
result.to_csv('scores_utr3.csv')
```
